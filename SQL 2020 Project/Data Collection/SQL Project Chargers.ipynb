Getting 100 chargers for each of the 10 cities I defined in 'cities_data.csv'

```python
import requests
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
cities_df = pd.read_csv('cities_data.csv', index_col=0)

lat_list = list(cities_df['Latitude'])
long_list = list(cities_df['Longitude'])
rad_list = list(cities_df['Radius'])
```

```python
api_url = 'https://api.openchargemap.io/v3/poi/'
api_key = '104aaf04-45b4-4997-8a4e-90e563c1dc7b'

charger_data = {
    'city_id':[],
    'latitude':[],
    'longitude':[],
    'usage_cost':[]
}

for i in range(10):
    params = {
        'key':api_key, 
        'countrycode':'US', 
        'latitude':lat_list[i], 
        'longitude':long_list[i], 
        'distance':rad_list[i],
        'maxresults':100
    }
    chargers_request = requests.get(api_url, params=params)
    chargers = json.loads(chargers_request.text)
    
    for charger in chargers:
        charger_data['city_id'].append(i+1)
        charger_data['latitude'].append(charger['AddressInfo']['Latitude'])
        charger_data['longitude'].append(charger['AddressInfo']['Longitude'])
        charger_data['usage_cost'].append(charger['UsageCost'])
```
