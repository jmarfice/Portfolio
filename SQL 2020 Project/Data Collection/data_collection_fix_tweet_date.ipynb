```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('mysql+mysqldb://USERNAME:PASSWORD@HOST/DATABASE?charset=UTF8')
```

```python
df = pd.read_sql('select * from tweets;', engine)
```

```python
df.head()
```

```python
date_list = df['tweet_date'][0].split(' ')
```

```python
date_list
```

```python
month_dict = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06', 'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
fixed_date = []

for i in range(len(df['tweet_id'])):
    date_list = df['tweet_date'][i].split(' ')
    year = date_list[5]
    month = month_dict[date_list[1]]
    day = date_list[2]
    time = date_list[3]
    fixed_date.append(year + '-' + month + '-' + day + ' ' + time)


```

```python
fixed_date
```

```python
df['fixed_date'] = fixed_date
```

```python
df.to_sql('tweets', engine, if_exists='append', index=False)
```
